```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MultiLabelBinarizer



# Preprocess the data
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df['Genres']), columns=mlb.classes_, index=df.index))
df = df.join(pd.DataFrame(mlb.fit_transform(df['Cast']), columns=[f"Actor{i}" for i in range(1, 6)], index=df.index))

# Define the features and target
features = ['Budget', 'Ratings'] + list(mlb.classes_) + [f"Actor{i}" for i in range(1, 6)]
target = 'Revenue'



# Split the data into training and testing sets
train = df.sample(frac=0.8, random_state=42)
test = df.drop(train.index)

# Train a random forest regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train[features], train[target])

# Make predictions on the test set
test['Predicted_Revenue'] = rf.predict(test[features])

# Evaluate the model's performance
mae = abs(test['Predicted_Revenue'] - test['Revenue']).mean()
print(f"Mean Absolute Error: {mae:.2f}")
```
